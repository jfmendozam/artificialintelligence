```python
import numpy as np

img = np.zeros((512, 512, 3))
# Imagen, coord-inicial, coord-final, BGR, grosor
cv2.rectangle(img, (200, 100), (300, 1), (255, 0, 255), -1)
cv2.imshow("new window", img)
cv2.waitKey(10000)
cv2.destroyAllWindows()
```

```python
# https://github.com/opencv/opencv/tree/master/data/haarcascades
# Download: https://github.com/opencv/opencv/blob/master/data/haarcascades/haarcascade_frontalface_default.xml
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
img = cv2.imread("images/mi_foto.jpg", 1)
faces = face_cascade.detectMultiScale(img, 1.1, 5)

for x, y, w, h in faces:
    print(x, y, w, h)
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 3)

cv2.imshow("colombiano", img)
cv2.waitKey(10000)
cv2.destroyAllWindows()
```
