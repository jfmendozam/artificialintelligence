```python
def read_text_file(filename):
    file = open(filename, "r", encoding="utf-8") 
    return file.read()
```

```python
text = read_text_file("quijote.txt")
print(text)
```

# Divide text into words using NLTK

```python
import nltk
words_nltk = nltk.tokenize.word_tokenize(text, language='spanish')
print(words_nltk)
print(len(words_nltk))
```

# Custom tokenization

```python
from nltk.tokenize import MWETokenizer
tokenizer = MWETokenizer([('Ingeniería', 'de', 'Sistemas')])
tokenizer.add_mwe(('Facultad', 'de', 'Ingeniería', 'de', 'Sistemas'))
tokens = tokenizer.tokenize('La Facultad de Ingeniería de Sistemas es una unidad académica de la Universidad Santo Tomás de Tunja.'.split())
print(tokens)
tokens = tokenizer.tokenize('Los egresados de Ingeniería de Sistemas tienen muchas oportunidades laborales.'.split())
print(tokens)
```

# Divide text into words using spaCy

```python
import spacy
nlp = spacy.load("es_core_news_sm")
doc = nlp(text)
words_spacy = [token.text for token in doc]
print(words_spacy)
print(len(words_spacy))
```

# Find the difference between the lists

```python
print(set(words_spacy)-set(words_nltk))
```

# Compare the time it takes using both methods

```python
import time

def split_into_words_nltk(text):
    words = nltk.tokenize.word_tokenize(text, language='spanish')
    return words

def split_into_words_spacy(text):
    doc = nlp(text)
    words = [token.text for token in doc]
    return words

start = time.time()
split_into_words_nltk(text)
print(f"NLTK: {time.time() - start} s")

start = time.time()
split_into_words_spacy(text)
print(f"spaCy: {time.time() - start} s")
```
