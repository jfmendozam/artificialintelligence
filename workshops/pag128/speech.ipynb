```python
import spacy
import nltk
#!python -m spacy download es_core_news_sm
#!python -m spacy download es_core_news_lg
small_model = spacy.load("es_core_news_sm")
large_model = spacy.load("es_core_news_lg")
```

```python
def read_text_file(filename):
    file = open(filename, "r", encoding="utf-8") 
    return file.read()
```

```python
def pos_tag_spacy(text, model):
    doc = model(text)
    words = [token.text for token in doc]
    pos = [token.pos_ for token in doc]
    return list(zip(words, pos))
```

```python
text = read_text_file("quijote.txt")
words_with_pos = pos_tag_spacy(text, small_model)
print(words_with_pos)
```

# Tokenization

```python
#nltk.download('averaged_perceptron_tagger')
#nltk.download('punkt')
#nltk.download('cess_esp')
def word_tokenize_nltk(text):
    return nltk.tokenize.word_tokenize(text, language='spanish')

def word_tokenize_spacy(text, model):
    doc = model(text)
    return [token.text for token in doc]
```

# Part of speech tagging using NLTK

```python
def pos_tag_nltk(text):
    words = word_tokenize_nltk(text)
    words_with_pos = nltk.pos_tag(words)
    return words_with_pos
```

```python
words_with_pos = pos_tag_nltk(text)
print(words_with_pos)
```
